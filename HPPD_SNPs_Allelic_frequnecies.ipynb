```python
## Choice of country samples (exemple from Burkina Faso)
sample_sets = ["AG1000G-BF-A", "AG1000G-BF-B", "AG1000G-BF-C"]
```

```python
## Choice of cohorts
cohorts = "cohort_admin1_year"
```

```python
## Call SNP frequencies in the DataFram (exemple for gambiae)
taxon_snp_allele_freqs_df = ag3.snp_allele_frequencies(
    transcript=transcript,
    cohorts=cohorts,
    sample_sets=sample_sets,
    sample_query= "taxon == 'gambiae'",
    min_cohort_size= 0
)
taxon_snp_allele_freqs_df
```

```python
## Statistics view on SNP
taxon_snp_allele_freqs_df.query("max_af > 0").groupby(["effect", "impact"]).size()
```

```python
# Visualising SNP allele frequencies with the heatmaps
```

```python
## Choice of samples
sample_sets = ["AG1000G-AO", "AG1000G-BF-A", "AG1000G-BF-B","AG1000G-BF-C", "AG1000G-CD", "AG1000G-CF", "AG1000G-CI", "AG1000G-CM-A", "AG1000G-CM-B", "AG1000G-CM-C", "AG1000G-FR", "AG1000G-GA-A", "AG1000G-GH", "AG1000G-GM-A", "AG1000G-GM-B", "AG1000G-GM-C", "AG1000G-GN-A", "AG1000G-GN-B", "AG1000G-GQ", "AG1000G-GW", "AG1000G-KE", "AG1000G-ML-A", "AG1000G-ML-B", "AG1000G-MW", "AG1000G-MZ", "AG1000G-TZ", "AG1000G-UG"]
```

```python
## Choice of cohorts
cohorts = "country"
```

```python
## Call all SNP frequencies in the DataFram
snp_allele_freqs_df = ag3.snp_allele_frequencies(
    transcript=transcript,
    cohorts=cohorts,
    sample_sets=sample_sets,
    drop_invariant=False,
)
snp_allele_freqs_df
```

```python
## Filters out non-synonymous SNPs with frequencies above 5%.
ns_snps_df = snp_allele_freqs_df.query("effect == 'NON_SYNONYMOUS_CODING' and max_af >= 0.05")
ns_snps_df
```

```python
## Country all taxon heatmap
ag3.plot_frequencies_heatmap(ns_snps_df)
```

```python
## Country heatmap by taxon
### Country by arabiensis
aa_arab_freqs_df = ag3.aa_allele_frequencies(
    transcript=transcript,
    cohorts=cohorts,
    sample_sets="3.0",
    sample_query="taxon == 'arabiensis'",
)
aa_arab_freqs_df = aa_arab_freqs_df.query("max_af > 0.05")
### Country by gambiae
aa_gamb_freqs_df = ag3.aa_allele_frequencies(
    transcript=transcript,
    cohorts=cohorts,
    sample_sets="3.0",
    sample_query="taxon == 'gambiae'",
)
aa_gamb_freqs_df = aa_gamb_freqs_df.query("max_af > 0.05")
### Country by coluzzii
aa_col_freqs_df = ag3.aa_allele_frequencies(
    transcript=transcript,
    cohorts=cohorts,
    sample_sets="3.0",
    sample_query="taxon == 'coluzzii'",
)
aa_col_freqs_df = aa_col_freqs_df.query("max_af > 0.05")
### Print for heatmap of the three taxons
print("arabiensis_frequencies_heatmap")
ag3.plot_frequencies_heatmap(aa_arab_freqs_df)
print("gambiae_frequencies_heatmap")
ag3.plot_frequencies_heatmap(aa_gamb_freqs_df)
print("coluzzi_frequencies_heatmap")
ag3.plot_frequencies_heatmap(aa_col_freqs_df)
```
